```python
#code to add column for day of year
import pandas as pd
import numpy as np
flights = pd.read_csv('flightsbyWW.csv')
flights = pd.DataFrame(flights)
```

```python
flights.head()
```

```python
flights2 = flights[['DAY_OF_WEEK','AIRLINE', 'FLIGHT_NUMBER', 'TAIL_NUMBER', 'ORIGIN_AIRPORT', 'DESTINATION_AIRPORT', 'ARRIVAL_DELAY', 'AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY', 'WW']].copy()
```

```python
flights2['DAY_COUNT'] = 0
flights2.head()
```

```python
import time
tim = time.time()
day = 1
for i,row in flights2.iterrows():
    flights2.set_value(i,'DAY_COUNT',day)
    if flights2.loc[i,'DAY_OF_WEEK'] != flights2.loc[i+1,'DAY_OF_WEEK']:
        day = day + 1
    else:
        continue
print "Time to run code is ", (time.time() - tim)/60, "minutes"
```
