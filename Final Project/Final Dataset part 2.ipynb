```python
import pandas as pd
import numpy as np

flights = pd.read_csv('final1.csv')
flights = pd.DataFrame(flights)
```

```python
flights1 = pd.DataFrame(flights1)
```

```python
flights1['DELAY'] = ""
```

```python
import time
tim = time.time()
for i,row in flights1.iterrows():
    if row['ARRIVAL_DELAY'] < 1:
        flights1.set_value(i,'DELAY',"On Time")
    else:
        if row['ARRIVAL_DELAY'] >= 1:
            if row['ARRIVAL_DELAY'] < 30:
                flights1.set_value(i,'DELAY',"1-30")
            else:
                if row['ARRIVAL_DELAY'] >= 30:
                    if row['ARRIVAL_DELAY'] < 60:
                        flights1.set_value(i,'DELAY',"31-60")
                    else:
                        flights1.set_value(i,'DELAY',"Frustratingly delayed!")

print "the time to run the code is", (time.time() - tim)/60, "minutes"
```
