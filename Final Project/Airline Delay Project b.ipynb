```python
#objective - Can Month of travel be used to predict whether the passenger will experience a weather delay or not?
import pandas as pd
import csv
import matplotlib.pyplot as plt
flight = pd.read_csv('Project/flights.csv')
flights = pd.DataFrame(flight)
flight.head(20)
```

```python
flights.shape
```

```python
wea_delay = flights.copy()
```

```python
wea_delay = wea_delay[pd.notnull(wea_delay['WEATHER_DELAY'])] 
```

```python
wea_delay['DELAY'] = 0
wea_delay.head()
```

```python
var = 0
for i,row in wea_delay.iterrows():
    if row['WEATHER_DELAY'] != 0.0:
        var = 1
        wea_delay.set_value(i,'DELAY',var)
    else:
        var = 0
        wea_delay.set_value(i,'DELAY',var)
```

```python
wea_delay.head(50)
```

```python
wea_delay['MONTH'].unique()
```

```python
dummy = pd.get_dummies(wea_delay['MONTH'], prefix='MONTH')
```

```python
dummy.head()
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
lm = LogisticRegression()
```

```python
lm.fit(dummy[['MONTH_1', 'MONTH_2', 'MONTH_3', 'MONTH_4', 'MONTH_5', 'MONTH_7', 'MONTH_8', 'MONTH_9', 'MONTH_10', 'MONTH_11', 'MONTH_12']], wea_delay['DELAY'])
```

```python
print lm.coef_
print lm.intercept_
```

```python
np.exp(0.40)
```

```python
#there is a 49% higher chance of a weather delay in February compared to June, all other parameters being the same
```

```python
#trying the decision tree model
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators = 20)

X = dummy[['MONTH_1', 'MONTH_2', 'MONTH_3', 'MONTH_4', 'MONTH_5', 'MONTH_7', 'MONTH_8', 'MONTH_9', 'MONTH_10', 'MONTH_11', 'MONTH_12']]
y = wea_delay['DELAY']
#X.drop('DELAY', axis=1, inplace=True)
    
    
```

```python
model.fit(X, y)
```

```python
model.score(X,y)
```

```python
#AUC for random forest classifier
from sklearn.cross_validation import cross_val_score

scores = cross_val_score(model, X, y, scoring='roc_auc', cv=5)
print('CV AUC {}, Average AUC {}'.format(scores, scores.mean()))
```

```python
wea_delay['PREDICTED_DELAY'] = ""
```

```python
wea_delay['PREDICTED_DELAY'] =  model.predict(X)
```

```python
wea_delay['PREDICTED_DELAY'].value_counts()
```

```python
wea_delay.head()
```

```python
from sklearn.metrics import classification_report
y_true = wea_delay['DELAY']
y_pred = wea_delay['PREDICTED_DELAY']
print classification_report(y_true, y_pred)
#ends up being a crappy model
```
